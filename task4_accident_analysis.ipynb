```python
import pandas as pd
import numpy as np

# Create synthetic accident dataset
np.random.seed(42)

n = 500

df = pd.DataFrame({
    'Hour': np.random.randint(0, 24, n),
    'Weather': np.random.choice(['Clear', 'Rain', 'Fog', 'Snow'], n),
    'Road_Condition': np.random.choice(['Dry', 'Wet', 'Icy'], n),
    'Severity': np.random.choice(['Low', 'Medium', 'High'], n)
})

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Hour'], bins=24)
plt.title("Accidents by Time of Day")
plt.show()
```

```python
sns.countplot(x='Weather', data=df)
plt.title("Accidents by Weather")
plt.show()
```

```python
sns.countplot(x='Road_Condition', data=df)
plt.title("Road Conditions")
plt.show()
```

```python
sns.countplot(x='Severity', data=df)
plt.title("Accident Severity")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```
